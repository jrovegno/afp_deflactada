```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
%matplotlib inline
```

```python
# Opciones escrituda de archivos csv compatibles con excel spanish
karg_csv = dict(delimiter=';', decimal=',', index_col=0, parse_dates=True)
```

```python
afps = ['CUPRUM', 'HABITAT', 'PLANVITAL', 'PROVIDA', 'MODELO']
```

```python
for afp_name in afps:
    # Descarga datos
    filename = 'https://raw.githubusercontent.com/collabmarket/data_afp/master/data/PAT-%s.csv'%afp_name
    pat = pd.read_csv(filename, **karg_csv)
    filename = 'https://raw.githubusercontent.com/collabmarket/data_afp/master/data/VC-%s.csv'%afp_name
    vc = pd.read_csv(filename, **karg_csv)
    # Grafica
    fig, axes = plt.subplots(ncols=2, figsize=(16,8))
    pat['2000':].plot(title="Patrimonio Fondos administrados por %s"%afp_name, ax=axes[0])
    vc['2000':].plot(title="Valores Cuota administrados por %s"%afp_name, ax=axes[1])
    plt.savefig('pat-vc/pat-vc-%s.png'%afp_name)
    plt.show()    
```

```python
for afp_name in afps:
    # Descarga datos
    filename = 'https://raw.githubusercontent.com/collabmarket/data_afp/master/data/PAT-%s.csv'%afp_name
    pat = pd.read_csv(filename, **karg_csv)
    filename = 'https://raw.githubusercontent.com/collabmarket/data_afp/master/data/VC-%s.csv'%afp_name
    vc = pd.read_csv(filename, **karg_csv)
    # Grafica
    fig, axes = plt.subplots(ncols=2, figsize=(16,8))
    pat['2005':].plot(title="Patrimonio Fondos administrados por %s"%afp_name, ax=axes[0])
    vc['2005':].pct_change().cumsum().plot(title="Rentabilidad Acumulada administrados por %s"%afp_name, ax=axes[1])
    plt.savefig('pat-rent/pat-rent-%s.png'%afp_name)
    plt.show()    
```
